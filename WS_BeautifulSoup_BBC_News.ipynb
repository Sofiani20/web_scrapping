## Importation de ces packages

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

## Saisie de l'URL sur lequel on veut travailler

```python
url="http://feeds.bbci.co.uk./news/rss.xml"
```

## Importation du code de la page

```python
response = requests.get(url) # envoie d'une requete vers l'url qu'on stocke dans la variable response
soup= BeautifulSoup(response.text,"html.parser")
soup # contient l'information brute du site
```

## Nous allons extraire les informations intéressantes
**Ces informations sont surtout dans les items**

```python
items=soup.find_all("item") # enregistre tous les item contenu sur la page
```

```python
items # c'est une liste contenant les items: <item> ... </item>
```

```python
items[0] # premier item 
```

#### on va manipuler ce premier item

```python
item_1 = items[0]
```

##### affichons le titre de ce item

```python
item_1.title.text
```

##### affichons la description du item

```python
item_1.description.text
```

## Création d'une liste contenant toutes les informations extraites de la liste items

```python
highlight_items=[]
```

```python
for i in items:
    news_i={}
    news_i['title']=i.title.text
    news_i['description']=i.description.text
    news_i['pubdate']=i.pubdate.text
    highlight_items.append(news_i)
```

```python
highlight_items
```

## Conversion de la liste en base de données

```python
df_items = pd.DataFrame(highlight_items,columns=['title','description','pubdate'])
```

```python
df_items
```

## Générer un fichier csv pour cette base de données

```python
df_items.to_csv("BBC_web_scrapping.csv",index=False,encoding='utf-8')
```
